```python
from skimage import io
```

```python
image_RGB = io.imread(r'../TheBasic/dataset/spiderman_low.jpg')
```

```python
io.imshow(image_RGB)
```

```python
image_RGB.shape
```

```python
from skimage import color
```

```python
image_XYZ = color.rgb2xyz(image_RGB)
```

```python
io.imshow(image_XYZ)
```

```python
from skimage import exposure
from pylab import *
```

```python
adjust_gamma1 = exposure.adjust_gamma(image_RGB, 0.3)
adjust_gamma2 = exposure.adjust_gamma(image_RGB, 3)
```

```python
figure(0)
io.imshow(adjust_gamma1)
figure(1)
io.imshow(adjust_gamma2)
```

```python
io.imsave('spiderman_gamma.jpg', adjust_gamma1)
```

```python
import numpy as np
```

```python
from skimage.metrics import structural_similarity as ssim
```

```python
np.squeeze(image_RGB).shape
```

```python
image_GRAY = color.rgb2gray(image_RGB)
image_GRAY.shape
```

```python
image_GRAY_gamma1 = color.rgb2gray(adjust_gamma1)
```

```python
SSIM_check_similarity  = ssim(image_GRAY, image_GRAY)
```

```python
# if near to 1, it's means close similar
SSIM_check_similarity
```

```python
SSIM_check_similarity2 = ssim(image_GRAY, image_GRAY_gamma1)
```
