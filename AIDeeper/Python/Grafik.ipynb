```python
# % Parameter
L = 1.0;         # % Panjang kotak
V0 = 0.0;        # % Potensial di dalam kotak
E = 1.0;         # % Energi partikel
m = 1.0;         # % Massa partikel
h_bar = 1.0;     # % Konstanta Planck tereduksi
```

```python
# % Hitung konstanta normalisasi
k = np.sqrt(2 * m * E / h_bar**2)

# % Tentukan jumlah langkah dan interval spatial
N = 1000         # % Jumlah langkah
dx = L / N       # % Interval spatial

# % Inisialisasi array
x = np.linspace(0, L, N+1)  # % Grid spasial
psi = np.zeros(N+1)      # % Fungsi gelombang
```

```python
# % Inisialisasi kondisi awal
psi[0] = 0;              # % Di dinding kiri
psi[N] = 0;            # % Di dinding kanan
psi[1] = 1;              # % Misalnya, di posisi awal
```

```python

# % Implementasi metode beda hingga
# for i = 2:N
#     psi(i+1) = 2 * psi(i) - psi(i-1) - 2 * dx^2 * (E - V0) * psi(i);
# end

for i in range(2, N+1):
    psi[i] = 2 * psi[i-1] - psi[i-2] - 2 * dx**2 * (E - V0) * psi[i-1]

# % Normalisasi fungsi gelombang
# psi_normalized = psi / sqrt(sum(abs(psi).^2 * dx));
psi_normalized = psi / np.sqrt(np.sum(np.abs(psi)**2) * dx)

# % Hitung probabilitas
# probability = abs(psi_normalized).^2;
probability = np.abs(psi_normalized)**2

# % Plot probabilitas
# figure;
# plot(x, probability);
# xlabel('Posisi (x)');
# ylabel('Probabilitas');
# title('Probabilitas Partikel dalam Kotak Kuantum');
plt.figure()
plt.plot(x, probability)
plt.xlabel('Posisi (x)')
plt.ylabel('Probabilitas')
plt.title('Probabilitas Partikel dalam Kotak Kuantum')
plt.show()


# % Tampilkan hasil
# fprintf('Probabilitas total: %f\n', sum(probability) * dx);
print(f'Probabilitas total: {probability}')
```
